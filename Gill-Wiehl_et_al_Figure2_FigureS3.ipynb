```python
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import fiona
# From GeoPandas, our world map data
worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
import pandas as pd
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
df = pd.read_excel('gillwiehl_et_al_data_map.xlsx')
df

df_all=pd.read_excel('gillwiehl_et_al_data_map_all.xlsx')
df_all

df_all_results=pd.read_excel('df_all_results_table.xlsx')
df_all_except_aus_results=pd.read_excel('df_all_except_aus_results_table.xlsx')
df_adoption_results=pd.read_excel('df_adoption_results_table.xlsx')
df_usage_results=pd.read_excel('df_usage_results_table.xlsx')
df_stacking_results=pd.read_excel('df_stacking_results_table.xlsx')
df_fNRB_results=pd.read_excel('df_fNRB_results_table.xlsx')
df_firewood_charcoal_results=pd.read_excel('df_only_charcoal_results_table.xlsx')
df_only_EF_results=pd.read_excel('df_EF_only_results_table.xlsx')
df_adjusted_results=pd.read_excel('df_adjusted_results_table.xlsx')
df_rebound_results=pd.read_excel('df_rebound_results_table.xlsx')

df_over_credit=pd.DataFrame(df_all_results.iloc[9:34,4], columns=['name'])
df_over_credit['name']=df_all_results.iloc[9:34,4]
df_over_credit['over_credit']=df_all_results.iloc[9:34,2]
df_over_credit['over_credit_all_but']=df_all_except_aus_results.iloc[9:34,2]
df_over_credit['over_credit_adopt']=df_adoption_results.iloc[9:34,2]
df_over_credit['over_credit_C']=df_usage_results.iloc[9:34,2]
df_over_credit['over_credit_stacking']=df_stacking_results.iloc[9:34,2]
df_over_credit['over_credit_B']=df_fNRB_results.iloc[9:34,2]
df_over_credit['over_credit_E']=df_firewood_charcoal_results.iloc[9:34,2]
df_over_credit['over_credit_D']=df_only_EF_results.iloc[9:34,2]
df_over_credit['over_credit_F']=df_adjusted_results.iloc[9:34,2]
df_over_credit['over_credit_G']=df_rebound_results.iloc[9:34,2]
df_over_credit

df=pd.merge(worldmap, df, how='left', on='name')
df_all=pd.merge(worldmap,df_all, how='left',on='name')
df_over_credit=pd.merge(worldmap,df_over_credit, how='left',on='name')
```

```python
fig = plt.figure(figsize =(7.08, 8.26))
ax1 = fig.add_subplot(2, 1, 1)
df_all.plot(color="white", ax=ax1, edgecolor="darkgrey")
df_all.plot(column='credits_issued', ax=ax1, legend=True,cmap='YlGn', legend_kwds={'label': "Issued Credits",
                        'orientation': "vertical"},vmin=0, vmax=10*1*10**5)
ax1.set_title('Main cookstove methodologies on the Voluntary Carbon Market')

ax0 = fig.add_subplot(2, 1, 2)
df.plot(color="white", ax=ax0, edgecolor="darkgrey")
df.plot(column='credits_issued', ax=ax0, legend=True,cmap='YlGn',legend_kwds={'label': "Issued Credits",
                        'orientation': "vertical"},vmin=0, vmax=10*1*10**5) #PuBu_r
ax0.set_title('Across our sample')

fig.tight_layout()


fig.savefig('Figure_2.png',dpi=600)
```
